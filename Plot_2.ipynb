```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
with open('HMN_history.pickle', 'rb') as handle:
    hmn = pickle.load(handle)
with open('fusion_history.pickle', 'rb') as handle:
    fusion = pickle.load(handle)
with open('CoA_Res_history.pickle', 'rb') as handle:
    res = pickle.load(handle)
with open('CoA_Res_history_2.pickle', 'rb') as handle:
    res2 = pickle.load(handle)
with open('DrQA_history.pickle', 'rb') as handle:
    drqa = pickle.load(handle)
```

```python
# summarize history for accuracy  
plt.figure(1)  
plt.plot(drqa['activation_20_acc']) 
plt.plot(fusion['start_acc']) 
plt.plot(hmn['activation_84_acc'])  
plt.plot(res['output_1_acc'])  
# plt.plot(a['val_start_acc'])  
plt.title('Training Accuracy')  
plt.ylabel('Accuracy')  
plt.xlabel('Epoch')  
plt.legend(['DrQA', 'Fusion', 'CoA-HMN', 'CoA-Res'], loc='upper left')  
plt.show()    
# summarize history for loss  
plt.figure(2)  
plt.plot(drqa['loss']) 
plt.plot(fusion['loss']) 
plt.plot(hmn['loss'])  
plt.plot(res['loss'])  
plt.title('Training Loss')  
plt.ylabel('Loss')  
plt.xlabel('Epoch')  
plt.legend(['DrQA', 'Fusion', 'CoA-HMN', 'CoA-Res'], loc='upper right')  
plt.show() 
```

```python
# summarize history for accuracy  
plt.figure(1)  
plt.plot(drqa['val_activation_20_acc']) 
plt.plot(fusion['val_start_acc']) 
plt.plot(hmn['val_activation_84_acc'])  
plt.plot(res['val_output_1_acc'])  
# plt.plot(a['val_start_acc'])  
plt.title('Validation Accuracy')  
plt.ylabel('Accuracy')  
plt.xlabel('Epoch')  
plt.legend(['DrQA', 'Fusion', 'CoA-HMN', 'CoA-Res'], loc='upper left')  
plt.show()    
# summarize history for loss  
plt.figure(2)  
plt.plot(drqa['val_loss']) 
plt.plot(fusion['val_loss']) 
plt.plot(hmn['val_loss'])  
plt.plot(res['val_loss'])  
plt.title('Validation Loss')  
plt.ylabel('Loss')  
plt.xlabel('Epoch')  
plt.legend(['DrQA', 'Fusion', 'CoA-HMN', 'CoA-Res'], loc='upper right')  
plt.show() 
```
